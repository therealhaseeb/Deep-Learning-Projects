```
import pandas as pd
import numpy as np
```

```
data = pd.read_csv('pima-indians-diabetes.csv')
```

```
data
```

```
data = data.values
data
```

```
x = data[:,0:8].astype(float)

from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
print(x_scaled)

x_train = x_scaled[0:540]
x_test = x_scaled[541:760]
```

```
y = data[:,8]
print(y)
y_train = y[0:540]
y_test = y[541:760]
```

```
len(x_train)
```

```
len(x_test)
```

```
from keras import models
from keras import layers
```

```
network = models.Sequential()
network.add(layers.Dense(12, activation='relu', input_dim=8))
network.add(layers.Dense(10, activation='relu'))
network.add(layers.Dense(8, activation='relu'))
network.add(layers.Dense(4, activation='relu'))
network.add(layers.Dense(1, activation='sigmoid'))
```

```
network.compile(optimizer='rmsprop',
                loss='binary_crossentropy',
                metrics=['accuracy'])
```

```
network.fit(x_train, y_train, validation_split=0.3, shuffle=True, epochs=70, batch_size=10)
```

```

network.evaluate(x_test, y_test)
```

```
vectors_answers = network.predict(x_test[210:])
vectors_answers = np.where(vectors_answers > 0.5, 1, 0)
print(vectors_answers)
```
